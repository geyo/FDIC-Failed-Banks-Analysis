```python
import pandas as pd
```

```python
# Files to load 
balance_sheet_to_load = "../source_data/BalanceSheetSummaries.csv"
bank_data_to_load = "../source_data/bank-data.csv"
banklist_data_to_load = "../source_data/banklist.csv"
```

```python
# Read files into and store into Pandas DataFrames
bank_data = pd.read_csv(bank_data_to_load, encoding='ANSI')
banklist_data = pd.read_csv(banklist_data_to_load, encoding='ANSI')
balance_sheets_data = pd.read_csv(balance_sheet_to_load, encoding='ANSI')
```

```python
# Remove extraneous columns
balance_sheets_data = balance_sheets_data.drop(['Year','Quarter','Cash and Investments', 'Due from FDIC Corp and Receivables',\
                                                'Assets in Liquidation', 'Estimated Loss on Assets in Liquidation(1)'], axis=1)
```

```python
balance_sheets_data.columns
```

```python
# Add up total liabilities
balance_sheets_data['Total_Liabilities'] = balance_sheets_data['Administrative Liabilities'] + balance_sheets_data['FDIC Subrogated Deposit Claim']\
                                         + balance_sheets_data['Uninsured Deposit Claims'] + balance_sheets_data['Other Claimant Liabilities']\
                                         + balance_sheets_data['Unproven Claims']
```

```python
# Create new DF with just columns needed. 
balance_sheets_data_clean = balance_sheets_data[['Fund Code', 'Receivership', 'Failure Date', 'Total Assets','Total_Liabilities', \
                                                 'Administrative Liabilities', 'FDIC Subrogated Deposit Claim', 'Uninsured Deposit Claims', \
                                                 'Other Claimant Liabilities', 'Unproven Claims','Net Worth (Deficit)']]
balance_sheets_data_clean
```

```python
balance_sheets_data_clean = balance_sheets_data_clean.iloc[:-3,:]
balance_sheets_data_clean
```

```python
balance_sheets_data_clean.to_csv('../static/data/balance_sheets_for_database.csv', index=False)
```
