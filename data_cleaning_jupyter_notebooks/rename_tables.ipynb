```python
import pandas as pd
```

```python
# Files to load 
bank_data = "../source_data/bank-data.csv"
updated_banks_with_coords = "../static/data/updated_banks_with_coordinates_qbfasset.csv"
balance_sheets_for_database = "../static/data/balance_sheets_for_database.csv"
trendchart = "../static/data/trendchart.csv"
banklist_for_dropdowndemo = "../static/data/banklist_for_dropdowndemo.csv"
banklist = "../source_data/banklist.csv"
```

```python
# Read the bank data CSV file into a pandas DataFrame
bank_data_df = pd.read_csv(bank_data, encoding='ANSI')
updated_banks_with_coords_df= pd.read_csv(updated_banks_with_coords, encoding='ANSI')
balance_sheets_for_database_df= pd.read_csv(balance_sheets_for_database, encoding='ANSI')
trendchart_df= pd.read_csv(trendchart, encoding='ANSI')
banklist_for_dropdowndemo_df= pd.read_csv(banklist_for_dropdowndemo, encoding='ANSI')
banklist_df= pd.read_csv(banklist, encoding='ANSI')

bank_data_df.head()
```

```python
#rename columns
bank_data_df.rename(columns={'CERT': 'cert',
                                'CHCLASS1': 'chclass1',
                                'CITYST': 'cityst', 
                                'COST': 'cost',
                                'FAILDATE': 'faildate',
                                'FIN': 'fund',
                                'ID': 'id',
                                'NAME': 'name',
                                'QBFASSET': 'qbfasset',
                                'QBFDEP': 'qbfdep',
                                'RESTYPE': 'restype',
                                'RESTYPE1': 'restype1',
                                'SAVR': 'savr'}, inplace=True, errors='raise')

bank_data_df.head()
```

```python
updated_banks_with_coords_df.head()
```

```python
#rename columns
updated_banks_with_coords_df.rename(columns={'Fund': 'fund',
                            'Location': 'location',
                            'Latitude': 'latitude', 
                            'Longitude': 'longitude',
                            'FAILDATE': 'faildate',
                            'QBFASSET': 'qbfasset'
                            }, inplace=True, errors='raise')

updated_banks_with_coords_df.head()
```

```python
#rename columns
balance_sheets_for_database_df.rename(columns={'Fund Code': 'fund',
                            'Receivership': 'receivership',
                            'Failure Date': 'faildate', 
                            'Total Assets': 'total_assets',
                            'Total_Liabilities': 'total_liabilities',
                            'Administrative Liabilities': 'administrative_liabilities',
                            'FDIC Subrogated Deposit Claim': 'fdic_subrogated_deposit_claim',
                            'Uninsured Deposit Claims': 'uninsured_deposit_claims', 
                            'Other Claimant Liabilities': 'other_claimant_liabilities',
                            'Unproven Claims': 'unproven_claims', 
                            'Net Worth (Deficit)': 'net_worth_deficit'                
                            }, inplace=True, errors='raise')

balance_sheets_for_database_df.head()
```
