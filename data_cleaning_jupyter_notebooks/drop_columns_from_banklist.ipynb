```python
import pandas as pd
```

```python
# Files to load 
banklist = "../source_data/banklist.csv"
```

```python
# Read the bank data CSV file into a pandas DataFrame
banklist_df = pd.read_csv(banklist, encoding='ANSI')

banklist_df.head()
```

```python
banklist_df = banklist_df.drop(index=0)

banklist_df.head()
```

```python
banklist_df = banklist_df.drop(index=1)

banklist_df.head()
```

```python
# Save the updated DataFrame to a new CSV file
banklist_df.to_csv('../source_data/banklist.csv', index=False)
```
