```python
import pandas as pd
from geopy.geocoders import Nominatim
```

```python
# File to load 
banklocations_data_to_load = "../source_data/banklocations.csv"
```

```python
# Read the CSV file into a pandas DataFrame
df = pd.read_csv(banklocations_data_to_load, encoding='ANSI')
df.head()
```

```python
# Initialize the Nominatim geocoder
geolocator = Nominatim(user_agent='my-app', timeout=10)

# Create two new columns to store the latitude and longitude values
df['Latitude'] = ''
df['Longitude'] = ''

# Loop through each row in the DataFrame
for i, row in df.iterrows():
    
    # Combine the city and state into a single string
    location = geolocator.geocode(row['Location'])
    if location is not None:
        
        # If the location is found, get the latitude and longitude
        df.at[i, 'Latitude'] = location.latitude
        df.at[i, 'Longitude'] = location.longitude
    
# # Save the updated DataFrame to a new CSV file
# df.to_csv('banks_with_coordinates.csv', index=False)
```

```python
# File to load 
bank_data_to_load = "bank-data.csv"
```

```python
# Read the bank data CSV file into a pandas DataFrame
df2 = pd.read_csv(bank_data_to_load, encoding='ANSI')
df2.head()
```

```python
# Rename FIN to Fund
df2 = df2.rename(columns={'FIN': "Fund"})
df2.head()
```

```python
# add qbfasset to bank
merged_df = pd.merge(df, df2, on="Fund")
merged_df
```
